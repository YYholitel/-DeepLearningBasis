```python
A = torch.arange(12,dtype= torch.float32).reshape((3,4)) # 创造了 二维数组 
```

```python
B = torch.tensor([[2,1,4,3],[1,2,3,4],[4,3,2,1]])
```

```python
torch.cat((A,B),dim = 0 ) , torch.cat((A,B) , dim =1) # 索引n-1 
```

```python
torch.cat ((A,B) ,dim = -1)  # 2-  1 的 索引
```

```python
A == B 
```
