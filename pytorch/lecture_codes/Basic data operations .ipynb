```python
#数据预处理

# 文件的操作

# 创建文件
# 创建文件一般使用os库 ， 关键操作在于makedirs，提供拼接的参数，
# 修改文件 使用 with open (data_file,
import os
os.makedirs(os.path.join('...','data'),exist_ok=True) # exist_ok防止重复报错
data_file = os.path.join(os.path.join('...','data'),'data.csv')
with open(data_file,'w') as f: # w代表write , f是简写
    f.write('a,b,c\n')
    f.write('a,b,c\n')
    f.write('a,b,NA\n')

# 读取文件 ，一般使用  pandas库存
import pandas as pd
data =pd.read_csv(data_file)
print(data)
data # 自动使用html实现
```

```python
# 对数字插值/删除
inputs, outputs = data.iloc[:, 0 :2],data.iloc[:,3] # index loaction ,访问指定数据的地址
inputs = inputs.fillna(inputs.mean())
# 填充 数字型下的NA位置并使用 均值替代
print(inputs)
```

```python
# 对类别值的操作
inputs = pd.get_dummies(inputs,dummy_na = True,dtype= int) # 把缺失值当成 独立类别处理
print(inputs)
```
