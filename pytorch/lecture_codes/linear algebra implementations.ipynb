```python
# 矩阵的自动求导
import torch
x = torch.arange(4.0)
x.requires_grad_(True) #显示梯度的存储
x.grad # 求解梯度
y = 2 * torch.dot(x,x)
y.backward()
x.grad
x.grad == 4 * x
```

```python
x.grad.zero_()# 梯度清零
u = y.detach() # 把y当成x的常数
z= u * x
z.sum().backward()
x.grad == u

```
